```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/Standard_Metropolitan_Areas_Data-data.csv")
df.head()
```

```python
# consecutive plots
plt.plot(df.work_force, df.income, color="r")
plt.show()

plt.plot(df.physicians, df.income)
plt.show()
```

```python
# by this way, 2 plots will appear on the same graph
plt.plot(df.work_force, df.income, color = "r")
plt.plot(df.physicians, df.income)

plt.show()
```

```python
# add title, labels and legend
plt.scatter(df.percent_senior, df.crime_rate)

# add title
plt.title("Crime Rate Against Percent Senior")

# add x and y label
plt.xlabel("Percent Senior")
plt.ylabel("Crime Rate")

plt.show()
```

```python
# add legend for mulltiplots in a graph
plt.plot(df.work_force, df.income, color="r", label="work_force")
plt.plot(df.physicians, df.income, label="physicians")

# legend
plt.legend()

plt.show()
```

```python
# changing color of plot
plt.plot(df.work_force, df.income, color="r", label="work_force")
plt.plot(df.physicians, df.income, color = "c", label="physicians")
plt.plot(df.hospital_beds, df.income, color = "0.75", label = "hospital_beds")

plt.legend()
plt.show()
```

```python
# linestyle: dashed("--"), solid("-"), dashdot("-."), dotted(":")
plt.plot(df.work_force, df.income, linestyle="--", color="r", label="work_force")
plt.plot(df.physicians, df.income, linestyle="-", color = "c", label="physicians")
plt.plot(df.land_area, df.income, linestyle="-.", color = "black", label = "land_area")
plt.plot(df.percent_city, df.income, linestyle=":", color="blue", label="percent_city")

plt.legend()
plt.show()
```

```python
# marker: circular("o"), cross("x"), triangles("^")
plt.plot(df.work_force, df.income, marker="o", color="r", label="work_force")
plt.plot(df.physicians, df.income, marker="x", color = "c", label="physicians")
plt.plot(df.hospital_beds, df.income, marker="^", color = "0.75", label = "hospital_beds")

plt.legend()
plt.show()
```

```python
# changing plot size
# method used: plt.figure(figsize=(width_px, height_px))
plt.plot(df.work_force, df.income)
plt.plot(df.physicians, df.income)
plt.plot(df.hospital_beds, df.income)

plt.figure(figsize=(20,16))
plt.show()
```

```python
# multiplot
# method: plt.subplot(n_rows, n_cols, index_of_plot)

plt.subplot(1, 2, 1)
plt.plot(df.work_force, df.income, "go")
plt.title("Income Against Work Force")

plt.subplot(1, 2, 2)
plt.plot(df.physicians, df.income, "ro")
plt.title("Income Against Physicians")

# main title of the whole figure
plt.suptitle("General Relations Betwenn Factor X and Income")
plt.figure(figsize=(16,5))
plt.show()
```

```python
# another alternative for multiplot indexing
fig, (ax1, ax2) = plt.subplots(1,2)

fig.suptitle("General Relations Betwenn Factor X and Income")

ax1.plot(df.work_force, df.income, "go")
ax1.set_title("Income Against Work Force")

ax2.plot(df.physicians, df.income, "ro")
ax2.set_title("Income Against Physicians")

plt.show()
```

# **Multiplot**

# plt.subplots()
  - Config `fig` and all items
  - All axes are stored in `ax` (array), acessed by ax[0], ax[1]...

```python
import seaborn as sns 

# 2 graphs in one row
# size of ax = n_cols
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
sns.histplot(data=df, x=df.columns[0],  ax=ax[0])
sns.boxplot(data=df, x=df.columns[0], ax=ax[1])
```

```python
# mutltiplot graph for all features, such that there're 2 features in a row
for idx in range(0,df.columns.shape[0]-1):
  fig, ax = plt.subplots(1, 2, figsize=(10,4))
  sns.histplot(data=df, x=df.columns[idx], ax=ax[0])
  sns.histplot(data=df, x=df.columns[idx+1], ax=ax[1])
```

# **plt.subplot()**
  - config each axes individually

```python
# config the figure size
plt.figure(figsize=(10,4)) 

# 2 features in a row
# plt.subplot(nrows, ncols, n_item)
ax1 = plt.subplot(1, 2, 1)
sns.histplot(data=df, x=df.columns[0], ax = ax1)

ax2 = plt.subplot(1, 2, 2)
sns.histplot(data=df, x=df.columns[1], ax=ax2)
```

```python
# multiplot of all features

n_plots = df.columns.shape[0]
n_cols = 2
n_rows = (n_plots+1) // n_cols

plt.figure(figsize=(20, 16))
for i, col in enumerate (df.columns):
  ax = plt.subplot(n_rows, 2, i+1)
  sns.boxplot(data=df, x=col, ax=ax)

plt.tight_layout()
```

```python
n_cols = 2
n_plots = df.columns.shape[0]
n_rows = (n_plots+1) // n_cols

plt.figure(figsize=(16,20))
for i in range(n_plots):
  ax = plt.subplot(n_rows, n_cols, i+1)
  sns.histplot(data=df, x=df.columns[i], ax=ax)

plt.tight_layout()
```
